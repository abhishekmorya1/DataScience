```python
import json
```

```python
# load the data

def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)

    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
# clean and structure the data

def clean_data(data):
    text_to_num = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5
    }

    clean_data = []
    unique_users = set()

    for user in data:

        # Clean the rating
        raw_rating = str(user.get("rating", "")).strip().lower()

        if raw_rating in text_to_num:
            user["rating"] = text_to_num[raw_rating]
        else:
            try:
                user["rating"] = float(raw_rating)
            except:
                user["rating"] = None

        # Handle missing age
        raw_age = user.get("age", "").strip()

        if raw_age == "":
            user["age"] = None
        else:
            user["age"] = int(raw_age)

        # Clean feedback
        user["feedback"] = user.get("feedback", "").strip().title()

        # Remove duplicate users
        if user["name"].strip() in unique_users:
            continue

        unique_users.add(user["name"].strip())
        clean_data.append(user)

    return clean_data
```

```python
data = clean_data(data)
```

```python
# get meaningful insights from data 

def get_insights(data):

    # Average Rating
    total_rating = 0

    for user in data:
        total_rating += float(user["rating"])

    print(f"The average rating is {total_rating / len(data)}")

    # percentage of users with poor ratings 
    poor_rating = 0
    for user in data:
        if float(user["rating"])< 3:
            poor_rating+=1

    print(f"percentage of users with poor rating = {poor_rating/len(data)*100}%")
    
```

```python
get_insights(data)
```

```python
# build a recommendation feature (Rule based, not ML)

def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]

        if float(user["rating"]) >= 4:
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"

        recommendations.append(curr_recomm)

    return recommendations
```

```python
get_recommendations(data)
```
