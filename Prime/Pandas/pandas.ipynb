```python
data2 = pd.read_csv("emp1 (4).csv")
data2
```

```python
data2.dropna()

# dropna() removes rows that contain missing values (NaN).

# how="any" (Default)
```

```python
data2.dropna(how="all")

# Remove a row only if ALL the values in that row are NaN
# If a row has at least one non-NaN value, it will not be removed
```

```python
data2.dropna(axis=1, how="all")

# This removes only those columns where every value is NaN.


# | Code                              | Meaning                                           |
# | --------------------------------- | ------------------------------------------------- |
# | `data2.dropna()`                  | Remove rows with **any** `NaN` (default behavior) |
# | `data2.dropna(how="any")`         | Same as above                                     |
# | `data2.dropna(how="all")`         | Remove rows only if **all** values are `NaN`      |
# | `data2.dropna(axis=1, how="all")` | Remove columns where all values are `NaN`         |
```

```python
data2.dropna(subset="ename")

# What does subset do?

# The subset parameter tells Pandas:

# "Check only the specified column(s) for missing values."

# Here, Pandas checks only the ename column.

# If ename contains NaN, that row is removed.
# It ignores NaN values in all other columns.
```

```python
# Check Multiple Columns
# You can also check more than one column.

data2.dropna(subset=["ename", "city"])
```

```python
data2.dropna(axis=0)

# Remove rows that contain missing (NaN) values.
# What does axis=0 mean?

# In Pandas:

# axis=0 → Operate on rows (default)
# axis=1 → Operate on columns
```

```python
data2.dropna(axis=1)

# Removes columns containing NaN
```
