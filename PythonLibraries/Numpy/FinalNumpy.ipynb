```python
# np.nditer()

# nditer stands for N-Dimensional Iterator.

# It iterates through every element of the array, regardless of its dimensions.

x=np.random.randint(1,30,(3,3))
x

# np.nditer() always returns one element at a time.
```

```python
for i in np.nditer(x):
    print(i)
```

```python
# np.ndenumerate()

# ndenumerate() returns both the index and the value of each element.

x=np.random.randint(1,30,(3,3))
x
```

```python
for index , value in np.ndenumerate(x):
    print(index, "->",value)
```

```python
x=np.random.randint(1,100,(2,3,3))
x
```

```python
for i , j in np.ndenumerate(x):
  print(i,"->",j)
```

```python
x=np.random.randint(1,200,(3,3))
x
```

```python
for i in range(len(x)):
   for j in range(len(x[i])):
      print(i,",",j, "->",x[i][j])
```

```python
# find sum of diagonal of matrix

x=np.random.randint(1,30,(3,3))
x                  
```

```python
sum=0
for i, j in np.ndenumerate(x):
    if i[0]==i[1]:
        sum=sum+j
print(sum)
        
```
