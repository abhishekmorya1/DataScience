```python
# copy vs view
# list uses copy and numpy uses view of data

list1 = [1,2,3,4,5]
sub_list = list1[1:3]
print(sub_list)
sub_list[0] = 500
print(sub_list)
print(list1)

# # numpy array
# npy = np.array([1,2,3,4,5])
# sub_array = npy[1:3]
# print(sub_array)
# sub_array[0] = 500
# print(sub_array)
# print(npy)

# numpy array
npy = np.array([1,2,3,4,5])
sub_array = npy[1:3].copy()
print(sub_array)
sub_array[0] = 500
print(sub_array)
print(npy)
```

```python
# What is Broadcasting?
# Broadcasting is a NumPy mechanism that allows arithmetic operations on arrays
# with different shapes by automatically expanding the smaller array.

# Why is Broadcasting useful?
# Less code
# Faster execution
# No manual loops
# Better performance

# In simple words:
# A small array is automatically expanded to match the shape of a larger array.
```

```python
# Difference Between hstack(), vstack(), and stack()
# Function	  Joins	                  Rule
# hstack()	Left to Right   	Same number of rows
# vstack()	Top to Bottom	    Same number of columns
# stack()	Creates a new axis	Arrays must have the same shape
```

```python

# Create two arrays
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

print("Array A:")
print(a)
print("Shape:", a.shape)

print("\nArray B:")
print(b)
print("Shape:", b.shape)

print("\n" + "="*50)

# 1. Horizontal Stack
print("1. HORIZONTAL STACK (hstack)")
print("Joins arrays Left to Right")
h = np.hstack((a, b))
print(h)
print("Shape:", h.shape)

print("\n" + "="*50)

# 2. Vertical Stack
print("2. VERTICAL STACK (vstack)")
print("Joins arrays Top to Bottom")
v = np.vstack((a, b))
print(v)
print("Shape:", v.shape)

print("\n" + "="*50)

# 3. Stack (axis = 0)
print("3. STACK (axis = 0)")
print("Creates a New Axis")
s0 = np.stack((a, b), axis=0)
print(s0)
print("Shape:", s0.shape)

print("\n" + "="*50)

# 4. Stack (axis = 1)
print("4. STACK (axis = 1)")
print("Creates a New Axis")
s1 = np.stack((a, b), axis=1)
print(s1)
print("Shape:", s1.shape)
```

```python
# hstack() and vstack() join arrays.
# stack() does not join on an existing axis; it creates a completely new axis, which is why the output becomes a higher-dimensional array (e.g., from 2D to 3D).
```

```python
# Difference Between These Functions
# Function	Works On	Splits
# split()	1D or 2D	Into equal parts or at given indices
# hsplit()	2D	       Columns (Left → Right)
# vsplit()	2D	       Rows (Top → Bottom)
```

```python
# np.unique()
# What is np.unique()?

# Definition:

# np.unique() returns only the unique (non-duplicate) values from an array.
```

```python

# ------------------------------
# Original Arrays
# ------------------------------

a = np.array([10,20,30,40,50,60])

b = np.array([[1,2,3,4],
              [5,6,7,8]])

c = np.array([[1,2],
              [3,4],
              [5,6],
              [7,8]])

print("="*50)
print("1. np.split()")
print("="*50)

print(a)
print()

print(np.split(a,3))

print("\n"+"="*50)
print("2. np.hsplit()")
print("="*50)

print(b)
print()

print(np.hsplit(b,2))

print("\n"+"="*50)
print("3. np.vsplit()")
print("="*50)

print(c)
print()

print(np.vsplit(c,2))
```

```python


a = np.array([10,20,30,10,20,40,50,40])

print("Original Array:")
print(a)

print("\n1. Unique Values")
print(np.unique(a))

print("\n2. Unique Values with Count")
values, counts = np.unique(a, return_counts=True)
print("Values :", values)
print("Counts :", counts)

print("\n3. First Occurrence Index")
values, index = np.unique(a, return_index=True)
print("Values :", values)
print("Index :", index)

print("\n4. Inverse Index")
values, inverse = np.unique(a, return_inverse=True)
print("Values :", values)
print("Inverse :", inverse)
```
