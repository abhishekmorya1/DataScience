**Operators in Python**
<!-- Operators are special symbols used to perform operations on variables and values. -->

```python
# <!-- 1. Arithmetic Operators -->
# <!-- Arithmetic operators are used to perform mathematical calculations. -->

a=10
b=4

print("Addition =", a+b) #addition
print("Subtration =",a-b) #subtraction
print("Multiplication =",a*b) #multiplication
print("Division =",a/b) # Divides first number by second
print("Floor Division =",a//b) #Returns only the integer part of division
print("Modulus =", a%b)  #Returns the remainder after division
print("Exponent =", a**b) #Raises a number to the power of another
```

```python
# 2. Comparison (Relational) Operators

# These operators compare two values and return True or False.

# | Operator | Meaning               |
# | -------- | --------------------- |
# | ==       | Equal to              |
# | !=       | Not equal             |
# | >        | Greater than          |
# | <        | Less than             |
# | >=       | Greater than or equal |
# | <=       | Less than or equal    |

x = 20
y = 15

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

```python
# 3. Assignment Operators
# These operators assign values to variables.

# | Operator | Example | Same As      |
# | -------- | ------- | ------------ |
# | =        | x = 5   | Assign value |
# | +=       | x += 3  | x = x + 3    |
# | -=       | x -= 3  | x = x - 3    |
# | *=       | x *= 3  | x = x * 3    |
# | /=       | x /= 3  | x = x / 3    |
# | //=      | x //= 3 | x = x // 3   |
# | %=       | x %= 3  | x = x % 3    |
# | **=      | x **= 3 | x = x ** 3   |


x = 10

x += 5
print(x)

x -= 2
print(x)

x *= 3
print(x)

x /= 2
print(x)
```

```python
# 4. Logical Operators
# Used to combine multiple conditions.

# | Operator | Meaning                             |
# | -------- | ----------------------------------- |
# | and      | Both conditions must be True        |
# | or       | At least one condition must be True |
# | not      | Reverses the result                 |

age = 22
# citizen = True

print(age >= 18 and citizen)
print(age < 18 or citizen)
print(not citizen)
```

```python
# 5. Bitwise Operators
# These operators work on the binary (bits) representation of integers.

# | Operator | Meaning     |
# | -------- | ----------- |
# | &        | AND         |
# | |        | OR          |
# | ^        | XOR         |
# | ~        | NOT         |
# | <<       | Left Shift  |
# | >>       | Right Shift |

a = 5
b = 3

print(a & b)
print(a | b)
print(a ^ b)
print(a << 1)
print(a >> 1)
```

```python
# 6. Membership Operators
# These operators check whether a value exists in a sequence such as a list, tuple, string, or dictionary.

# | Operator | Meaning     |
# | -------- | ----------- |
# | in       | Present     |
# | not in   | Not Present |


fruits = ["apple", "banana", "mango"]

print("apple" in fruits)
print("grapes" in fruits)
print("grapes" not in fruits)
```

```python
# 7. Identity Operators
# These compare whether two variables refer to the same object in memory, not just whether their values are equal.

# | Operator | Meaning           |
# | -------- | ----------------- |
# | is       | Same object       |
# | is not   | Different objects |

a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b) #comparing objects
print(a is c) #comparing objects

print(a == c) #comparing values
```
