```python
# find even number between any range using for loop

for i in range(1, 20):
    if i%2==0:
        print(i)
```

```python
# Create a particular table using for loop
x=5
for i in range(1,11):
    print(x, "x", i, "=", x*i)
```

```python
# dynamic table using for loop like 4-8

for i in range(4,9):
    print("Table of:",i)
    for j in range(1,11):
        print(i , "x", j, "=", i*j)
    print()
```

```python
# sum of n natural number

sum = 0
for i in range(1,6):
    sum+= i
print("Sum of first 5 natural numbers is:", sum)
```

```python
# find the numbers between 1 to 100 which is divide by 3 and 4

count=0
for i in range(1,100):
    if i%3==0 and i%4==0:
        # print(i)
        count+=1
print("The count is:" , count)
```

```python
# prime number using for loop 

num = 26
count = 0

for i in range(1 , num+1):
    if num%i==0:
        count+=1

if(count == 2):
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
# print prime number between 1 to 100 using for loop

for i in range(1, 101):
    count = 0
    for j in range(1, i+1):
        if i%j==0:
            count+=1
    if(count == 2):
        print(i)
```

```python
# fibonacci series between range

for i in range(10):
    a = 0
    b = 1
    if i <= 1:
        print(i)
    else:
        c = a + b
        a = b
        b = c
print(c)
```
