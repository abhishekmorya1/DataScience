```python
# extend
# extend() is a list method that adds all elements of one list to another list.

# list1 → The list that will be modified.
# list2 → The list whose elements will be added to list1.

x1 = [1,2,3]
x2 = [4,5,6]
```

```python
x1.extend(x2)
```

```python
x1
```

```python
# Comparison Table
# Method	         Output
# x1.append(x2) 	[1, 2, 3, [4, 5, 6]]
# x1.extend(x2)	    [1, 2, 3, 4, 5, 6]
```
