```python
# variables in Python
# A variable is a name used to store data in memory.
# can contain letter , numbers  and underscore ( can't start with numbers and can't use python keywords)

name = "Abhishek Morya"
age = 25
Education = "Post Graduation"

print(name)
print(age)
print(Education)

```

```python
# variable Assignment
x=1
y=2
print(x+y)
```

```python
# Multiple Variable Assignment

a,b,c = 10,20,100
print(a)
print(b)
print(c)
```

```python
# Same Value to Multiple Variables

x=y=z = 200
print(x)
print(y)
print(z)
```

```python
# primitive datatype

# int - 10, 20 , 30, -90, -99
# float - 10.0 , 23.5, 77.3 
# str - "abhishek" , "morya", 'python'

```

```python
# Dynamic Typing
# Python automatically detects the data type.

name = "abhishek"
print(type(name))

age = 25
print(type(age))
```

```python
# Naming Conventions (Best Practice)
# Use snake_case.

student_name = "abhishek"
student_education = "Post Graduation"
```

```python
# Deleting a Variable

d=10
# del d
print(d)
```

```python
# example

name = "Abhishek Morya"
course = "Python"
duration = "9 Months"

print("Name:", name)
print("Course", course)
print("Duration", duration)
```

```python
# Quick Interview Points

-> A variable stores data.
-> Assignment operator is =.
-> Python is dynamically typed.
-> Variable names are case-sensitive (age and Age are different).
-> Follow snake_case naming convention.
    
# Quick Revision
    
-> Variable = Container for storing data.
-> Syntax: variable_name = value
-> Use meaningful names.
-> Python automatically identifies the data type.
-> Check data type with type().
```
