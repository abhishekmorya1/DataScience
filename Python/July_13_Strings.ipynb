```python
# String 
# any character within quotes (single, double, triple) act as string
```

```python
s1= 'abhishek'
s2 = "abhishek"
s3 = '''abhishek'''
print(s1, type(s1))
print(s2, type(s2))
print(s3, type(s3))
```

```python
# difference between single, double and triple quotes 

# | Quote Type | Best Use                                                              |
# | ---------- | --------------------------------------------------------------------- |
# | `' '`      | Simple strings                                                        |
# | `" "`      | Strings containing apostrophes (`I'm`, `don't`)                       |
# | `''' '''`  | Multi-line strings and docstrings                                     |
# | `""" """`  | Multi-line strings and docstrings (most commonly used for docstrings) |


# Use single quotes when your string contains double quotes.

s1 = 'python' 
print(s1)

text = 'She said "Hello"'
print(text)

# Use double quotes when your string contains single quotes (apostrophes).

s2 = "My name is Abhishek Morya"
print(s2)

text2 = "I'm learning Python."
print(text2)


#  triple string Used for multi-line strings and docstrings.

message = """Hello,
Welcome to Python.
Have a great day!"""

print(message)

# Docstring Example
def add(a, b):
    """Returns the sum of two numbers."""
    return a + b

print(add.__doc__)
```

```python
# String supports Indexing

# String Indexing in Python
# String indexing means accessing individual characters of a string using their position (index).
# Python uses zero-based indexing, which means the first character starts at index 0.

text = "Python"

print(text[0])
print(text[1])
print(text[2])
print(text[5])

# Negative Indexing
# Negative indexing starts from the end of the string.

text1 = "Python"

print(text1[-1])
print(text1[-2])
print(text1[-6])


# Using len() with Indexing
# The len() function returns the number of characters in a string.

text3 = "Python"

print(len(text3))

# Index Out of Range
# If you try to access an index that doesn't exist, Python raises an error.

# text4 = "Python"

# print(text4[10])


# | Index Type      | Starts From   | Example                |
# | --------------- | ------------- | ---------------------- |
# | Positive Index  | Beginning     | `text[0]`, `text[2]`   |
# | Negative Index  | End           | `text[-1]`, `text[-2]` |
# | First Character | `0`           | `text[0]`              |
# | Last Character  | `-1`          | `text[-1]`             |
# | Invalid Index   | Outside range | `IndexError`           |

# Key Points

# Indexing is used to access individual characters in a string.
# Python uses 0-based indexing.
# Negative indexing accesses characters from the end.
# Strings are immutable, so you cannot modify characters using indexing.
```

```python
# Slicing in Python

# String slicing is used to extract a portion (substring) of a string.

# string[start : stop : step]
# start → Starting index (included)
# stop → Ending index (excluded)
# step → Number of positions to move (optional)

# Remember: The stop index is not included in the output.

# Basic Slicing
text = "Python"

print(text[0:2])
print(text[2:5])
print(text[1:4])

# Omitting Start Index -> If start is omitted, Python starts from index 0
text = "Python"

print(text[:3])

# Omitting Stop Index -> If stop is omitted, python goes till the eIndentationError

print(text[1:])

# Omitting Both Start and Stop -> print the copy of the sBytesWarning

print(text[:])

# using step -> The step tells Python how many positions to skip.

print(text[::2])

# Reverse a string -> Use a negative step.
print(text[::-1])

# Negative Indexing -> Negative Index Slicing

text2 = "Python"

print(text2[-4:-1])
print(text2[0:11:2])



text3 = "Programming"

print(text3[:7])      # Program
print(text3[3:])      # gramming
print(text3[::3])     # Pgmn
print(text3[-5:])     # mming
print(text3[:-3])     # Programm
print(text3[::-1])    # gnimmargorP
```

```python
# Common String Methods in Python

# Python provides many built-in methods to work with strings. Four of the most commonly used are:

# len() – Returns the length of the string.
# index() – Returns the position of a substring.
# count() – Counts how many times a substring appears.
# replace() – Replaces one substring with another.

text = "Python"

print(len(text)) 

text1 = "Hello World"

print(len(text1)) #space is also counted

# Index
# string.index(value, start, end)
# value → Substring to search for (required)
# start → Starting index (optional)
# end → Ending index (optional)
text2 = "Python"

print(text2.index("t"))

text3 = "Programming"

print(text3.index("g"))
print(text3.index("g", 4)) #using start


# Count Method ->The count() method returns the number of times a substring appears in the string.
# string.count(value, start, end)

text4 = "banana"

print(text4.count("a"))

text5 = "Python is easy. Python is powerful."

print(text5.count("Python"))

# Replace Method -> The replace() method replaces all occurrences of one substring with another.

# string.replace(old, new, count)
# old → Text to replace
# new → Replacement text
# count → Number of replacements (optional)

text6 = "Hello World"

print(text6.replace("World", "Python"))

text7 = "banana"

print(text7.replace("a", "o"))

text8 = "banana"

print(text8.replace("a", "o", 1))


text9 = "Programming"

print("Length :", len(text9))
print("Index of 'g' :", text9.index("g"))
print("Count of 'm' :", text9.count("m"))
print("Replace :", text9.replace("Programming", "Python"))
```
